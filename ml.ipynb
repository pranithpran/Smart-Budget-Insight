```python
from sklearn.preprocessing import LabelEncoder
```

```python
lb1 = LabelEncoder()
lb2 = LabelEncoder()
```

```python
df['Occupation'] = lb1.fit_transform(df['Occupation'])
df['City_Tier'] = lb2.fit_transform(df['City_Tier'])
```

```python
X = df[['City_Tier' ,'Occupation','Dependents','Income','Age','Loan_Repayment','Insurance','Rent']]
y = df['Education']
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.25,random_state=42)
```

```python
from sklearn.tree import DecisionTreeRegressor
from sklearn.linear_model import LinearRegression
```

```python
dtr = DecisionTreeRegressor()
dtr.fit(X_train, y_train)
```

```python
lr = LinearRegression()
```

```python
lr.fit(X_train,y_train)
```

```python
lr.score(X_test,y_test)
```

```python
dtr.score(X_test,y_test)
```

```python
from sklearn.model_selection import cross_val_score
```

```python
cross_val1 = cross_val_score(dtr,X,y,cv=5)
cross_val1
```

```python
cross_val2 = cross_val_score(lr,X,y,cv=5)
cross_val2
```

```python
import pickle

# Assuming 'model' is your trained machine learning model
with open("Education.pickle", "wb") as file:
    pickle.dump(lr, file)
```
